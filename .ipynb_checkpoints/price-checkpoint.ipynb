### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Datasets


```python
price = pd.read_csv('prices.csv')
```

```python
price.head()
price.shape
```

#### 1. Selecting and plotting GE, GOOG, MSFT, TSLA respectively by using seaborn pairplot.

#### GE

```python
ge = price[price['symbol'] == 'GE']
ge
```

```python
ge.describe()
```

```python
ge.set_index('date', inplace = True)
```

```python
ge[['open','close']].plot(figsize = (15,5))
```

```python
sns.pairplot(ge[['open','close']])
plt.title('GE stock exchange data')
plt.show()
```

#### GOOG

```python
goog = price[price['symbol'] == 'GOOG']   
```

```python
goog.describe()
```

```python
goog[['open','close']].plot(figsize = (15,5))
```

```python
sns.pairplot(goog[['open','close']])
plt.title('GOOG stock exchange price')
plt.show()
```

### MSFT

```python
concat = pd.concat([ge,goog,msft,tsla])
```

### Plotting of Stock Performance between jan 2015 to jan 2016

```python
price.set_index('date', inplace = True)
price
```

```python
lower_date = "2015-01-01"
upper_date = "2016-01-30"
date_filter = (price.index >lower_date) & (price.index <upper_date)
final_data = price.loc[date_filter]
```

```python
final_data.plot(figsize = (15,5))
plt.title('Plotting of Stock Performance between jan 2015 to jan 2016')
plt.show()
```

```python
final_data
```

```python
sns.pairplot(final_data)
plt.show()
```

### Calculating the  20 days, 50 days, and  90 days moving average of the above stocks

```python
twenty = price.rolling(window=20).mean()
twenty.head(20)
```

```python
fifty = price.rolling(window=50).mean()
fifty.tail(50)
```

```python
ninety = price.rolling(window = 90).mean()
ninety.tail(90)
```
